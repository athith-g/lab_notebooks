```python
ema_raw = pd.read_excel(path/'TEAM300_EMA_7.3.23.xlsx')
```

```python
ema_raw.fillna(-999, inplace=True)
```

```python
import datetime

def extract_b_for_sorting(row):
  return int(row["day"][1])
def extract_d_for_sorting(row):
  try:
    return int(row["day"][-2:])
  except:
    return int(row["day"][-1])

def move_sr_to_br(d):
  if d["Time"] >= datetime.time(21) and (d["Survey abbrev"] in ["SCMR", "SMRP", "SMRB"]):
    return "BR"
  return d["Survey abbrev"]

def apply_letter_code(d):
  if d["Survey abbrev"] == "MR":
    return "A"
  if d["Survey abbrev"] == "BR":
    return "C"
  return "B"

def move_date(d):
  if d["Date"] == datetime.date(2022, 3, 10):
    return datetime.date(2022, 3, 11)
  elif d["Date"] == datetime.date(2022, 11, 22):
    return datetime.date(2022, 11, 21)
  return d["Date"]
```

```python
ema_raw["Survey abbrev"] = ema_raw.apply(move_sr_to_br, axis=1)
ema_raw["letter_code"] = ema_raw.apply(apply_letter_code, axis=1)
ema_raw["Date"] = ema_raw.apply(move_date, axis=1)
ema_raw = ema_raw.rename(columns={"Date": "date"})
```

```python
ema_raw = ema_raw.merge(times[["date", "day"]], on="date")
```

```python
final = ema_raw[["SID", "day", "letter_code", "MJ", "Vaping"]].groupby(["SID", "day", "letter_code"]).max().reset_index()
```

```python
def generate_all_combos():
  map = { "SID": [], "day": [], "letter_code": []}
  for id in ids.PID.unique():
    for day in ema_raw.day.unique():
      for letter in ema_raw.letter_code.unique():
        map["SID"].append(id)
        map["day"].append(day)
        map["letter_code"].append(letter)
  return map
```

```python
combos = pd.DataFrame(generate_all_combos())
```

```python
last = combos.merge(final, on=["SID", "day", "letter_code"], how="left").fillna(-999)
```

```python
def make_binary(d):
  if d > 0:
    return 1
  return d
```

```python
last["MJ"] = last["MJ"].apply(make_binary)
last["Vaping"] = last["Vaping"].apply(make_binary)
```

```python
def has_either(row):
  if row["MJ"] == -999:
    return -999
  if row["MJ"] == 1 or row["Vaping"] == 1:
    return 1
  return 0
```

```python
last["either"] = last.apply(has_either, axis=1)
```

```python
last = last.rename(columns={"letter_code": "Survey"})
last = last[["SID", "Survey", "day", "MJ", "Vaping", "either"]]
```

```python
last["b"] = last.apply(extract_b_for_sorting, axis=1)
last["d"] = last.apply(extract_d_for_sorting, axis=1)
```

```python
last = last.sort_values(by=["SID", "b", "d", "Survey"]).drop(columns=["b", "d"])
```

```python
last.to_excel("revised_ema.xlsx", index=False)
```

```python
def widen_timeline(df, id_col_name, bw_name, col_codes):
  transpose_dict = {"id": df[id_col_name].unique().tolist()}

  for code in col_codes:
    for ID in df[id_col_name].unique():
      for BW in df[bw_name].unique():

        col_name = f"{col_codes[code]}_{BW}"
        if col_name not in transpose_dict:
          transpose_dict[col_name] = []

        row = df.loc[(df[id_col_name] == ID) & (df[bw_name] == BW)]
        transpose_dict[col_name].append(row.iloc[0][code])

  return pd.DataFrame(transpose_dict)
```

```python
t1
```

```python
parcel_calculations = {"PID": t1.PID.unique().tolist()}

def calculate_parcels(PID, code):
  for p in parcels:
    days = parcels[p]
    vals = []
    for d in days:
      val = t1.loc[(t1.PID == PID) & (t1.bwd == d)].iloc[0][code]
      vals.append(val)

    calculated_parcel = sum(vals)/len(vals)
    col_name = f"U_{p}_{codes[code]}"
    if col_name not in parcel_calculations:
      parcel_calculations[col_name] = []

    parcel_calculations[col_name].append(calculated_parcel)
```

```python
for PID in parcel_calculations["PID"]:
  for code in codes:
    calculate_parcels(PID, code)
```

```python
def calculate_week(df):
  burst = df["burst"]
  week = int(df["day"]) // 8 + 1
  return f"B{burst}W{week}"
```

```python
mlm["week"] = mlm.apply(calculate_week, axis=1)
```

```python
columns = mlm.columns.to_list()
variables_to_condense = columns[:2] + columns[5:-7] + columns[-2:]
variables_for_sum = columns[-7:-2]
```
