Timeline (I think that was weekly B last time) -

```python
def condense_bwd(timeline):
  return f"B{timeline['burst']}W{timeline['week']}D{timeline['day']}"
```

```python
def used_cannabis(timeline):
  return 1 if timeline['cannabis'] == 1 else timeline['cannabis']
```

```python
def used_ecig(timeline):
  return 1 if timeline['ecig'] == 1 else timeline['ecig']
```

```python
timeline["bwd"] = timeline.apply(condense_bwd, axis=1)
```

```python
timeline["bwd"].unique()
```

```python
d = {"PID":[], "bwd": []}
```

```python
for id in timeline["PID"].unique():
  for bwd in timeline["bwd"].unique():
    d["PID"].append(id)
    d["bwd"].append(bwd)
```

```python
timeline_complete = pd.DataFrame(d)
```

```python
timeline_complete = timeline_complete.merge(timeline, on=["PID", "bwd"], how="left")
```

```python
timeline_complete['cannabis_binary'] = timeline.apply(used_cannabis, axis=1)
```

```python
timeline_complete['ecig_binary'] = timeline.apply(used_ecig, axis=1)
```

```python
timeline_complete.fillna(-999, inplace=True)
```

```python
ema_raw = pd.read_excel('TEAM300_EMA_7.3.23.xlsx')
```

```python
ema_raw.fillna(0, inplace=True)
```

```python
"BR" in ema_raw.loc[ema_raw["SID"] == 101]["Survey abbrev"].unique()
```

```python
import datetime

def move_sr_to_br(d):
  if d["Time"] >= datetime.time(21) and (d["Survey abbrev"] in ["SCMR", "SMRP", "SMRB"]) and "BR" not in ema_raw.loc[ema_raw["SID"] == d["SID"]]["Survey abbrev"].unique():
    return "BR"
  return d["Survey abbrev"]

def apply_letter_code(d):
  if d["Survey abbrev"] == "MR":
    return "A"
  if d["Survey abbrev"] == "BR":
    return "C"
  return "B"

def move_date(d):
  if d["Date"] == datetime.date(2022, 3, 10):
    return datetime.date(2022, 3, 11)
  elif d["Date"] == datetime.date(2022, 11, 22):
    return datetime.date(2022, 11, 21)
  return d["Date"]
```

```python
ema_raw["Survey abbrev"] = ema_raw.apply(move_sr_to_br, axis=1)
ema_raw["letter_code"] = ema_raw.apply(apply_letter_code, axis=1)
ema_raw["Date"] = ema_raw.apply(move_date, axis=1)
ema_raw = ema_raw.rename(columns={"Date": "date"})
```

```python
ema_raw = ema_raw.merge(times[["date", "day"]], on="date")
```

```python
ema_raw
```

```python
final = ema_raw[["SID", "day", "letter_code", "MJ", "Vaping"]].groupby(["SID", "day", "letter_code"]).sum().reset_index()
```

```python
map = { "SID": [], "day": [], "letter_code": []}
```

```python
for id in ema_raw.SID.unique():
  for day in ema_raw.day.unique():
    for letter in ema_raw.letter_code.unique():
      map["SID"].append(id)
      map["day"].append(day)
      map["letter_code"].append(letter)
```

```python
combos = pd.DataFrame(map)
```

```python
last = combos.merge(final, on=["SID", "day", "letter_code"], how="left").fillna(-999)
```

```python
def make_binary(d):
  if d > 0:
    return 1
  return d
```

```python
last["MJ"] = last["MJ"].apply(make_binary)
last["Vaping"] = last["Vaping"].apply(make_binary)
```

```python
last = last.rename(columns={"letter_code": "Survey"})
last = last[["SID", "Survey", "day", "MJ", "Vaping"]]
```
