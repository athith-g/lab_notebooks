```python
import json
import pandas as pd
from pprint import pprint
```

```python
def process_responses(responses):
  videos = {}
  for response in responses:
    for video in response['data']:
      if video['type'] == 4:
        continue
      try:
        id = video['item']['id']
        likes = video['item']['stats']['diggCount']
        comments = video['item']['stats']['commentCount']
        views = video['item']['stats']['playCount']
        shares = video['item']['stats']['shareCount']
        desc = video['item']['desc']
        author = video['item']['author']['uniqueId']
        url = f"https://www.tiktok.com/@{author}/video/{id}"

        videos[id] = {
            'likes': likes,
            'comments': comments,
            'views': views,
            'shares': shares,
            'desc': desc,
            'url': url,
        }
      except KeyError:
        print(video)

  return pd.DataFrame(videos)
```

```python
responses = []

for i in range(10):
  r= input(f"{i+1}: Enter response: ")
  r.replace('true', 'True')
  r.replace('false', 'False')
  r = json.loads(r)
  responses.append(r)
```

```python
df = process_responses(responses)
df = df.T
```

```python
url = df.iloc[1]['url']
url
```

```python
csv = df.iloc[0:100].reset_index().rename(columns={'index': "video_id", 'desc': 'caption'})
csv
```

```python
def get_hashtags(df):
  d = df['caption']
  l = r = 0

  hashtags = []

  while l < len(d):
    if d[l] == "#":
      r = l+1
      while r < len(d) and d[r] != " " and d[r] != "#":
        r += 1

      hashtag = d[l:r]
      hashtags.append(hashtag)
      l = r-1

    l += 1

  return hashtags
```

```python
csv['hashtags_in_caption']  = csv.apply(get_hashtags, axis=1)
```

```python
csv.to_excel('whiskey.xlsx', index=False)
```
